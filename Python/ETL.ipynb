```python
def parse_xml(xml_file, df_cols):

    tree = ET.parse(xml_file)
    root = tree.getroot()
    rows = []

    for node in root:
        res = []
        for el in df_cols[:]:
            if node is not None and node.find(el) is not None:
                res.append(node.find(el).text)
            else:
                res.append(None)
        rows.append({df_cols[i]: res[i] for i, _ in enumerate(df_cols)})

    out_df = pd.DataFrame(rows, columns=df_cols)
    return out_df
```

```python
def extract_all_xml():

    path = '/Users/mike/IBM-Data-Engineering-Specialization-Coursera/IBM-Data-Engineering-Specialization-Coursera/Python/dealership_data'
    all_xml_files = Path(path).glob("*.xml")
    df_cols = ['car_model', 'year_of_manufacture', 'price', 'fuel']
    rows = list()

    for xmlfile in all_xml_files:

        tree = ET.parse(xmlfile)
        root = tree.getroot()
        
        for node in root:
            res = []
            for el in df_cols[:]:
                if node is not None and node.find(el) is not None:
                    res.append(node.find(el).text)
                else:
                    res.append(None)
                    
            rows.append({df_cols[i]: res[i] for i, _ in enumerate(df_cols)})
            
        out_df = pd.DataFrame(rows, columns=df_cols)

    return out_df
```

```python
def extract_all_csv_json():
    
    path = '/Users/mike/IBM-Data-Engineering-Specialization-Coursera/IBM-Data-Engineering-Specialization-Coursera/Python/dealership_data'
    all_csv = Path(path).glob("*.csv")
    all_json = Path(path).glob("*.json")
    dfs = []

    for csvfile in all_csv:
        df = pd.read_csv(csvfile)
        dfs.append(df)

    for jsonfile in all_json:
        df = pd.read_json(jsonfile, lines=True)
        dfs.append(df)

    
    out_df = pd.concat(dfs, axis=0, ignore_index=True)

    return out_df
```

```python
def extract():
    extracted_csv_json = extract_all_csv_json()
    extracted_xml = extract_all_xml()
    extracted_data = pd.concat([extracted_csv_json,extracted_xml], axis=0, ignore_index=True)
    return extracted_data
```

```python
def load(targetfile,data_to_load):
    data_to_load.to_csv(targetfile)
```

```python
def log(message):
    timestamp_format = '%Y-%h-%d-%H:%M:%S' # Year-Monthname-Day-Hour-Minute-Second
    now = datetime.now() # get current timestamp
    timestamp = now.strftime(timestamp_format)
    with open("logfile.txt","a") as f:
        f.write(timestamp + ',' + message + '\n')
```

```python
def transform(data):
        data['price'] = pd.to_numeric(data['price'])
        data['price'] = round(data['price'],2)
        return data
```

```python
log("ETL Job Started")
log("Extract phase Started")
extracted_data = extract()
log("Extract phase Ended")
extracted_data
log("Transform phase Started")
transformed_data = transform(extracted_data)
log("Transform phase Ended")
transformed_data
log("Load phase Started")
load(targetfile,transformed_data)
log("Load phase Ended")
log("ETL Job Ended")
```
